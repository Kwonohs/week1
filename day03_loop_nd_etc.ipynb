## 5.2 반복문

반복문은 어떤 수행 작업을 한 번이 아니라 계속해서 수행해야 할 때 사용.  

반복문은 while과 for 두 가지 존재.

### 5.2.1 while

1.   while  
while문은 특정 조건을 만족하는 동안 반복해서 수행해야 하는 코드를 작성할 때 사용.  

* while문의 기본 구조는 다음와 같음  
while 조건식:  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문
     
     조건식의 결과가 True이면 while문을 실행  
     반복문 실행이 종료되면 다시 조건식으로 이동  
     조건식의 결과가 False이면 더 이상 반복 실행문을 실행하지 않음
<br>

2.   while문이 필요한 경우  
while문은 주로 횟수가 정해지지 않은 연산을 진행할 때나  
어떤 특정 동작이 지속될 때 사용
<br>


3.   while문의 중첩
while문 내부에 또 다른 while문이 나타나는 것을 while문의 중첩이라고 함.  
반복처리해야 하는 대상이 2개 이상이면 while문의 중첩이 필요.  

     다음과 같은 예제서 사용  
     5일동안 매일 3시간씩 수업을 진행  
     매 시간 '1일차 1교시'와 같은 메세지를 출력  
     1일차부터 5일차까지 반복되는 일수와
     1교시부터 3교시까지 반복되는 시간이라는  
     2개의 반복처리 대상이 있으니 while문의 중첩 필요

```python
i = 1
while True:
    j = 1
    while True:
        print(f"{j} * {i} = {j * i}", end = '\t')
        j += 1
        if j == 10:
            break
    print()
    i += 1
    if i == 10:
        break
```

### 5.2.2 for

for문은 값의 범위나 횟수가 정해져 있을 때 사용.  
while문에 비해서 자주 사용되는 반복문.

* for문 기본 구조  
for 변수 in 반복가능객체:  
&nbsp;&nbsp;&nbsp;&nbsp;반복실행문

for문은 조건식이 존재하지 않고 '변수'와 '반복가능객체'만 사용하고 있음.  

**반복가능객체**  
여러 개의 데이터를 저장하고 있는 객체를 '반복가능객체'라고 부름  
for문에서 사용되는 반복가능한 주요 객체는 다음과 같음  

|구분|자료형|예제|
|-|-|-|
|sequence|문자열|'hello'|
||리스트|[1, 2, 3]|
||튜플|(1, 2, 3)|
||range|range(100)|
|non-sequence|set|{1, 2, 3}|
||tuple|{'a': 'apple', 'b': 'banana'}|


```python
# for 변수명 in collection, 문자열, range가 올 수 있다

# 변수 이름은
# snake 방식 -> hello_world
# upper camel 방식 -> HelloWorld

k = {1,2,3}
for i in k:
    print(i + 1)
```

```python
a = (1,2,3,4)
b = [1,2,3,4]
print(a[1:2])
print(b[1:2])

for i in "hello world":
    print(i)

for char in "hello world":
    print(char)
```

##### for문과 딕셔너리

for문과 딕셔너리를 활요하면 딕셔너리의 각 key, value을  
하나씩 꺼내어 사용 가능  

**사용법**  
1. key 요소만 접근할 경우  
for key in {딕셔너리}:
&nbsp;&nbsp;&nbsp;&nbsp;반복실행문  
그러나 명확한 전달을 위해 아래와 같이 사용을 권함

     for key in {딕셔너리}.keys():  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문  

2. value 요소만 접근할 경우  
for key in {딕셔너리}.values():  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문  


3. key, value 요소 모두 접근할 경우  
for key, value in {딕셔너리}.items():  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문  

```python
dictionary = {'a' : 'apple', 'b' : 'banana', 'c' : 'carrot'}

for i in dictionary:
    print(i) #default는 키가 반환


for i in dictionary.values():
    print(i)


for key, value in dictionary.items():
    print(key, value)
```

## 5.3 기타 제어문

### 5.3.1 break

break문은 while문이나 for문과 같은 반복문을
강제로 종료하고자 할 때 사용하는 제어문

반복문 내에서 break문이 나타나면 곧바로 break문이 포함된 반복문은 종료  

```python
total = 0
for i in range(1,11):
    if i % 5 == 0:
        break

    total += i

print(total)
```

```python
total = 0
for i in range(1,11):
    if i % 5 == 0:
        continue

    total += i

print(total)
```

### 5.3.2 continue

continue문은 반복문의 시작 지점으로 제어의 흐름을 옮기는 역할을 수행.  

만약 1에서 100부터의 합을 구한다고 할 때 3의 배수를  
합계에서 제외하고 싶은 경우 등에 사용

```python
def test():
    total = 0
    for i in range(1,11):
        if i % 5 == 0:
            return

        total += i
        print(total)

    print(total)
```

```python
test()
```

```python
total = 0
for i in range(0, 10, 2):
    if i % 3 != 0:
        total += i
print(total)
```

### 5.3.3 return

return은 함수 끝에 입력돼서 해당 함수를 호출한 곳으로   
데이터를 반환하는 역할 또는, 함수의 종료를 알리는 기능.

```python
user_input = int(input())
basket = []
for _ in range(user_input):
    fruit = input()
    if fruit not in basket:
        basket.append(fruit)


print(basket)
```

```python
user_input = input()

# 97 122 65 90
for char in user_input:
    if 97 <= ord(char) <= 122:
        print(char.upper(), sep = "", end = "")
    else:
        print(char.lower(), sep = "", end = "")

```
